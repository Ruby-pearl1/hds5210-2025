### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
# Identify numeric columns
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
numeric_cols
```

```python
import matplotlib.pyplot as plt

col = "total_medicare_allowed_amt"

plt.figure(figsize=(10,6))
plt.hist(df[col], bins=50)
plt.title(f"Distribution of {col}")
plt.xlabel(col)
plt.ylabel("Frequency")
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df[col], bins=50, log=True)
plt.title(f"Distribution of {col} (Log Scale)")
plt.xlabel(col)
plt.ylabel("Frequency (Log Scale)")
plt.grid(False)
plt.show()
```

To analyze the range and distribution of numeric values in my dataset, I created a histogram for the variable total_medicare_allowed_amt, which represents the total Medicare-allowed amount for each provider. This column was chosen because it has one of the highest levels of numeric diversity and provides meaningful insight into Medicare spending patterns.
I generated two histograms: one using a standard frequency scale and another using a logarithmic scale. The standard histogram showed that most providers fall into the lower range of allowed amounts, while the logarithmic version revealed a long right-skewed distribution, indicating that a smaller number of providers have extremely high Medicare-allowed totals.
These visualizations help illustrate how healthcare spending is distributed across providers and can support further analysis such as identifying outliers, segmenting providers, or normalizing data for modeling.
